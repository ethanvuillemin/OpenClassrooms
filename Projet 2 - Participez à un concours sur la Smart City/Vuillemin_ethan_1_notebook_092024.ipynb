## Etape 3 - RÃ©alisez une analyse univariÃ© de chaque variable du je de donnÃ©es

CommenÃ§ons avec un digramme de boite Ã  moustache de la hauteur et de la circonfÃ©rence des arbres

```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

# Exemple de donnÃ©es (remplace ceci par ton DataFrame)
# df = ...

# CrÃ©ation de la figure et des axes
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 6))

# CrÃ©ation du premier boxplot
sns.boxplot(data=df[['hauteur_m']], ax=ax1)
ax1.set_title('Hauteur en mÃ¨tres')
# dÃ©finir une Ã©chelle logarithmique
ax1.set_yscale('log')

# CrÃ©ation du deuxiÃ¨me boxplot
sns.boxplot(data=df[['circonference_cm']], ax=ax2)
ax2.set_title('CirconfÃ©rence en cm')
# dÃ©finir une Ã©chelle logarithmique
ax2.set_yscale('log')

# Ajuster les espacements
plt.tight_layout()

# Afficher les graphiques
plt.show()
```

Ensuite, nous allons afficher un diagramme circulaire des diffÃ©rents types d'arbres prÃ©sent.

```python
# Afficher les types d'arbres les plus prÃ©sent
category_counts = df['libelle_francais'].value_counts()
plt.pie(category_counts[:6], labels=category_counts.index[:6], autopct='%1.1f%%', startangle=90)
plt.show()       
```

Maintenant nous allons afficher le stade de developpement pour les especes les plus prÃ©sentes

```python
df_pieplot = df.copy()
# CrÃ©er un tableau de sous-graphes
fig, axs = plt.subplots(1, 3, figsize=(15, 5))

# Liste des types d'arbres
types_arbres = ['Platane', 'Marronnier', 'Tilleul']  # Remplace par les types que tu veux

for ax, arbre in zip(axs, types_arbres):
    arbre_pop = df[df['libelle_francais'] == arbre]
    category_counts = arbre_pop['stade_developpement'].value_counts()

    # CrÃ©er le diagramme ciculaire
    ax.pie(category_counts, labels=category_counts.index, autopct='%1.1f%%', startangle=90)
    ax.set_title(arbre)  # Ajouter un titre pour chaque sous-graphique

plt.tight_layout()  # Ajuster les espacements
plt.show()
```
