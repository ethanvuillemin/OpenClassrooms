## Etape 3 - Réalisez une analyse univarié de chaque variable du je de données

Commençons avec un digramme de boite à moustache de la hauteur et de la circonférence des arbres

```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

# Exemple de données (remplace ceci par ton DataFrame)
# df = ...

# Création de la figure et des axes
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 6))

# Création du premier boxplot
sns.boxplot(data=df[['hauteur_m']], ax=ax1)
ax1.set_title('Hauteur en mètres')
# définir une échelle logarithmique
ax1.set_yscale('log')

# Création du deuxième boxplot
sns.boxplot(data=df[['circonference_cm']], ax=ax2)
ax2.set_title('Circonférence en cm')
# définir une échelle logarithmique
ax2.set_yscale('log')

# Ajuster les espacements
plt.tight_layout()

# Afficher les graphiques
plt.show()
```

Ensuite, nous allons afficher un diagramme circulaire des différents types d'arbres présent.

```python
# Afficher les types d'arbres les plus présent
category_counts = df['libelle_francais'].value_counts()
plt.pie(category_counts[:6], labels=category_counts.index[:6], autopct='%1.1f%%', startangle=90)
plt.show()       
```

Maintenant nous allons afficher le stade de developpement pour les especes les plus présentes

```python
df_pieplot = df.copy()
# Créer un tableau de sous-graphes
fig, axs = plt.subplots(1, 3, figsize=(15, 5))

# Liste des types d'arbres
types_arbres = ['Platane', 'Marronnier', 'Tilleul']  # Remplace par les types que tu veux

for ax, arbre in zip(axs, types_arbres):
    arbre_pop = df[df['libelle_francais'] == arbre]
    category_counts = arbre_pop['stade_developpement'].value_counts()

    # Créer le diagramme ciculaire
    ax.pie(category_counts, labels=category_counts.index, autopct='%1.1f%%', startangle=90)
    ax.set_title(arbre)  # Ajouter un titre pour chaque sous-graphique

plt.tight_layout()  # Ajuster les espacements
plt.show()
```

Maintenant, que nous connaissons comment sont reparti les différentes espèces d'arbres, on peut regarder comment s'en occuper au mieux:

D'après le site [PictureThis](https://www.picturethisai.com/fr/) qui est un site d'aide botanique nous pouvons extraire ces informations:

- Pour [s'occuper d'un platane](https://www.picturethisai.com/fr/care/Platanus_hybrida.html#:~:text=Arrosez%20votre%20Platane%20commun%20deux,sol%20et%20r%C3%A9guler%20la%20temp%C3%A9rature.) il est conseillé de l'arroser une fois **toutes les 2 semaines** et conserver une** humidité moyenne** au possible

- Pour [s'occuper d'un marronier](https://www.picturethisai.com/fr/care/Aesculus_hippocastanum.html#:~:text=Arrosez%20votre%20Marronnier%20d'Inde,de%20chaleur%20et%20de%20s%C3%A9cheresse.) il est conseillé de l'arroser une fois **toutes les 1/2 semaines** et conserver une **humidité moyenne** au possible

- Pour [s'occuper d'un tilleul](https://www.picturethisai.com/fr/care/Tilia_platyphyllos.html) il est conseillé de l'arroser une fois **toutes les 2 semaines** et conserver une **humidité moyenne**au possible

À savoir que tous ces arbres sont définis comme étant très résistant a la sécheresse et que leurs entretiens sont facile à faire.
