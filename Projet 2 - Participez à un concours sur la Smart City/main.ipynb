
# Étape 4 - Détecter et traiter les valeurs aberrantes du jeu de données :



- Identifier les valeurs aberrantes dans le jeu de données, c'est-à-dire les valeurs qui s'écartent significativement de la distribution normale.

- Évaluer si ces valeurs aberrantes doivent être corrigées ou supprimées.

- Appliquer les techniques appropriées pour traiter ces valeurs aberrantes, telles que la suppression des valeurs extrêmes ou leur remplacement par des valeurs plus appropriées.




Sachant que l'abre le plus grand de paris est de 30m: https://agriculture.gouv.fr/lun-des-plus-grands-arbres-de-paris-veille-sur-le-78-rue-de-varenne

```python
# Methode baser sur les quantiles
q75_value = np.quantile(df['hauteur_m'], q=0.99)

df = df[df['hauteur_m'] <= q75_value]
print(f"Hauteur du plus grand arbre: {df['hauteur_m'].max()}m")
sns.boxplot(df['hauteur_m'])
```

```python
# Methode baser sur les quantiles
q75_value = np.quantile(df['circonference_cm'], q=0.99)

df = df[df['circonference_cm'] <= q75_value]
print(f"ciconference max: {df['circonference_cm'].max()}cm")
sns.boxplot(df['circonference_cm'])
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Comptage des types d'arbres par arrondissement
data = df['arrondissement'].value_counts()
data_df = data.reset_index()
data_df.columns = ['arrondissement', 'count']

# Création du barplot
sns.barplot(y='arrondissement', x='count', data=data_df)
plt.xticks(rotation=90)  # Rotation des labels pour plus de lisibilité
plt.title('Distribution des types d\'arbres par arrondissement')
plt.ylabel('Arrondissement')
plt.xlabel('Nombre d\'arbres')
plt.show()

```
